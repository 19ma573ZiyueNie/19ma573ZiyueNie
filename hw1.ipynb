$$ pdf:  \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2} {2\sigma^2} }=\frac{1}{2\sqrt{2\pi}} e^{-\frac{(x-1)^2} {8} } $$

```
print('--->>> Here is the graph of pdf. ')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def norm(x,mu,sigma):
    pdf = np.exp(-((x - mu)**2)/(2*sigma**2)) / (sigma * np.sqrt(2*np.pi))
    return pdf
x = np.linspace(-8, 10, 100)
plt.plot(x,norm(x,1,2))
```
